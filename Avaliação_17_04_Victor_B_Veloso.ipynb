#1 Par ou Ímpar

```python
#Entrada
a = int (input("digite um número"))

# Processamneto
resto = a % 2
# Saída
if resto == 0:
  print (f"O número {a} é par")
else:
  print (f"O número {a} é ímpar")
```

#2 Maior de dois números

#4 Classificação de idade


```python
#Entrada
idade = int(input("Digite a sua idade: "))

#Processamento e Classificação
if idade >= 60:
  print("Você é Idoso")
elif idade >= 18:
  print("Você é Adulto")
elif idade >= 13:
  print("Você é Adolescente")
elif idade >= 0:
  print("Você é Criança")
else:
  print("Idade inválida")
```

#6 Positivo, negativo ou zero

```python
#Entrada
numero = int(input("Digite a seu número: "))

#Processamento e Classificação
if numero >= 1:
  print("Seu número é positivo")
elif numero == 0:
  print("Seu número é zero")
else:
  print("Seu número é negativo")
```

#7 Calculadora simples

```python
a = int(input("digite o primeiro número: "))
b = int(input("digite o segundo número: "))

print ("Opção 1 = soma")
print ("Opção 2 = subtração")
print ("Opção 3 = multiplicação")
print ("Opção 4 = divisão")

opcao = int (input("Digite a opção desejada: "))

if opcao == 1:
  print (f"{a} + {b} = {a + b}")
elif opcao == 2:
  print (f"{a} - {b} = {a - b}")
elif opcao == 3:
  print (f"{a} * {b} = {a * b}")
elif opcao == 4:

  if b != 0:
    print (f"{a} / {b} = {a / b}")
  else:
    print("Erro: Divisão por zero não é permitida")
else:
  print ("Opção inválida")
```

#8 Dia da semana

```python
#Entrada
a = int(input("digite um número de 1 á 7: "))


if a == 1:
  print (" 1 = domingo")
elif a == 2:
  print (" 2 = segunda")
elif a == 3:
  print (" 3 = terça")
elif a == 4:
  print (" 4 = quarta")
elif a == 5:
  print (" 5 = quinta")
elif a == 6:
  print (" 6 = sexta")
elif a == 7:
  print (" 7 = sábado")

else:
  print ("Opção inválida")
```

#10 Desconto em compras

```python
#Entrada
valor_compra = float(input("Digite o valor da compra: "))

#Processamento
if valor_compra >= 100:
  desconto = 0.10 # 10%
elif valor_compra >= 50 and valor_compra < 100:
  desconto = 0.05  # 5%
else:
  desconto = 0.0   # Sem desconto

valor_desconto = valor_compra * desconto
valor_final = valor_compra - valor_desconto

#Saída
print(f"Valor da compra: R${valor_compra:.2f}")
if desconto > 0:
  print(f"Desconto aplicado: {desconto*100:.0f}%")
  print(f"Valor do desconto: R${valor_desconto:.2f}")
else:
  print("Nenhum desconto aplicado.")
print(f"Valor final a pagar: R${valor_final:.2f}")
```
