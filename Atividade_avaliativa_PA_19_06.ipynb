QUESTÃO 02 – Soma dos Números (2 pontos)

```python
#Solicite ao usuário um número inteiro positivo. Utilizando um laço de repetição, calcule e exiba a soma de todos os números de 1 até o número informado.

numero = int(input("Digite um número: "))
soma = 0
for i in range(1, numero + 1):
    soma += i
print(f"Resultado: {soma}")
```

QUESTÃO 03 – Tabuada (2 pontos)


```python
#Solicite ao usuário um número inteiro e exiba sua tabuada de 1 até 10.

numero = int(input("Digite um número para a tabuada: "))
print(f"\nTabuada do {numero}:")
for i in range(1, 11):
    print(f"{numero} x {i} = {numero * i}")
```

QUESTÃO 04 – Fatorial (4 pontos)

```python
#Solicite ao usuário um número inteiro positivo e calcule seu fatorial utilizando um laço de repetição.

numero = -1
while numero < 0:
    try:
        numero = int(input("Digite um número inteiro positivo para calcular o fatorial: "))
        if numero < 0:
            print("Erro: O número deve ser positivo. Tente novamente.")
    except ValueError:
        print("Erro: Entrada inválida. Por favor, digite um número inteiro.")

fatorial = 1
if numero == 0:
    print(f"Fatorial = 1") # 0! is 1
else:
    for i in range(1, numero + 1):
        fatorial *= i
    print(f"Fatorial = {fatorial}")
```

QUESTÃO 05 – Caixa Eletrônico (5 pontos)

```python

# Solicite ao usuário o valor do saque e informe a menor quantidade possível de notas para realizar o pagamento.

valor_saque = -1
while valor_saque < 0:
    try:
        valor_saque = int(input("Digite o valor do saque: "))
        if valor_saque < 0:
            print("Erro: O valor do saque deve ser positivo. Tente novamente.")
    except ValueError:
        print("Erro: Entrada inválida. Por favor, digite um número inteiro.")

valor_restante = valor_saque

print("\nNotas a serem entregues:")

# Notas de R$100
if valor_restante >= 100:
    quantidade_notas = 0
    while valor_restante >= 100:
        valor_restante -= 100
        quantidade_notas += 1
    print(f"{quantidade_notas} nota(s) de R$100")

# Notas de R$50
if valor_restante >= 50:
    quantidade_notas = 0
    while valor_restante >= 50:
        valor_restante -= 50
        quantidade_notas += 1
    print(f"{quantidade_notas} nota(s) de R$50")

# Notas de R$20
if valor_restante >= 20:
    quantidade_notas = 0
    while valor_restante >= 20:
        valor_restante -= 20
        quantidade_notas += 1
    print(f"{quantidade_notas} nota(s) de R$20")

# Notas de R$10
if valor_restante >= 10:
    quantidade_notas = 0
    while valor_restante >= 10:
        valor_restante -= 10
        quantidade_notas += 1
    print(f"{quantidade_notas} nota(s) de R$10")

# Notas de R$5
if valor_restante >= 5:
    quantidade_notas = 0
    while valor_restante >= 5:
        valor_restante -= 5
        quantidade_notas += 1
    print(f"{quantidade_notas} nota(s) de R$5")

# Notas de R$2
if valor_restante >= 2:
    quantidade_notas = 0
    while valor_restante >= 2:
        valor_restante -= 2
        quantidade_notas += 1
    print(f"{quantidade_notas} nota(s) de R$2")

# Notas de R$1
if valor_restante >= 1:
    quantidade_notas = 0
    while valor_restante >= 1:
        valor_restante -= 1
        quantidade_notas += 1
    print(f"{quantidade_notas} nota(s) de R$1")

if valor_saque == 0:
    print("Nenhuma nota a ser entregue para saque de R$0.")
```
