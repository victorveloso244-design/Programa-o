```python
#1 Contagem simples - o usuário digita um número e o programa imprimi do zero até o núme
num = int(input("Digite um número: "))
for i in range(num + 1):
  print(i)
```

```python
#2 Contagem Regressiva - o usuário digita um número e o programa imprimi o número até o zero
num = int(input("Digite um número: "))
for i in range(num, -1, -1):
  print(i)
```

```python
#3 Números Pares - o usuário digita um número e o programa imprimi os números pares
num = int(input("Digite um número: "))
for i in range(num + 1):
  if i % 2 == 0:
    print(i)
```

```python
#4 Números Impares - o usuário digita um número e o programa imprimi os números ímpares
num = int(input("Digite um número: "))
for i in range(num + 1):
  if i % 2 != 0:
    print(i)
```

```python
#5 Soma - o usuário digita um número e o programa soma os números do zero até o número
num = int(input("Digite um número: "))
soma = 0
for i in range(num + 1):
  soma += i
print(f"A soma dos números de 0 a {num} é: {soma}")
```

```python

#6 Tabuada - o usuário digita um número e o programa imprimi a tabuada desse número
num = int(input("Digite um número para a tabuada: "))
for i in range(1, 11):
  print(f"{num} x {i} = {num * i}")
```

```python
# 7 - Contar letras: O usuário digita uma palavra e o programa conta as letras da palavra.
palavra = input("Digite uma palavra: ")
numero_letras = len(palavra)
print(f"A palavra '{palavra}' tem {numero_letras} letras.")
```

```python

# 8 - Contar vogais: O usuário digita uma palavra e o programa conta as vogais.
palavra = input("Digite uma palavra: ")
vogais = "aeiouAEIOU"
contador_vogais = 0
for letra in palavra:
  if letra in vogais:
    contador_vogais += 1
print(f"A palavra '{palavra}' tem {contador_vogais} vogais.")
```

```python
# 9 - Maior número digitado: O usuário digita quatro números e o programa imprime o maior número digitado.
num1 = int(input("Digite o primeiro número: "))
num2 = int(input("Digite o segundo número: "))
num3 = int(input("Digite o terceiro número: "))
num4 = int(input("Digite o quarto número: "))

maior_numero = max(num1, num2, num3, num4)

print(f"O maior número digitado é: {maior_numero}")
```

```python
# 10 - O usuário digita um número entre 01 e 100 e o programa mostra quais são divisíveis por 3 e 5.
while True:
  num = int(input("Digite um número entre 1 e 100: "))
  if 1 <= num <= 100:
    break
  else:
    print("Número inválido. Por favor, digite um número entre 1 e 100.")

print(f"Números de 1 a {num} divisíveis por 3 e 5:")
for i in range(1, num + 1):
  if i % 3 == 0 and i % 5 == 0:
    print(i)
```
