Escrever um programa que recebe a temperatura em graus celsius e converte para Fahrenheit
#Entrada: Graus Celsius

#Processamento

F = C x (9/5) + 32

#Saída

Mostrar graus em Fahrenheit

```python
# Escrever um programa que soma dois números
# Entrada

a = int (input("digite o primeiro número"))
b = int (input("digite o segundo número"))

# Processamneto
soma= a + b
# Saída

print (f"O resultado da soma de {a} mais {b} é igual: {soma}")
```

```python
# Escrever um programa que faça cálculos aritméticos
# Entrada

a = int (input("digite o primeiro número"))
b = int (input("digite o segundo número"))

# Processamneto

soma= a + b
subtração= a - b
multiplicação= a*b
divisão= a/b
div_int= a//b
resto = a % b # o símbolo de % é igual ao resto da divisão

# Saída

print ("---------------CALCULADORA SIMPLES---------------")
print ("                                                 ")
print ("-----------------------soma------------------------")
print ("                                                 ")
print(f"O resultado da soma de {a} mais {b} é igual: {soma}")
print ("                                                  ")
print ("---------------------subtração------------------")
print ("                                                 ")
print (f"O resultado da subtração de {a} menos {b} é igual: {subtração}")
print ("                                                 ")
print ("---------------------multiplicação------------------")
print ("                                                 ")
print(f"O resultado da multiplicação de {a} multiplicado {b} é igual: {multiplicação}")
print ("                                                 ")
print ("---------------------Divisão------------------")
print ("                                                 ")
print (f"O resultado da divisao de {a} dividido {b} é igual: {divisão}")
print ("                                                 ")
print ("---------------------Divisão inteira------------------")
print ("                                                 ")
print (f"O resultado da divisao de {a} dividido {b} é igual: {div_int}")
print ("                                                 ")
print ("---------------------Resto------------------")
print ("                                                 ")
print (f"O resultado da divisao de {a} dividido {b} o resto é igual: {resto}")
print ("                                                 ")
```

```python
#Entrada
print("--------------Entrada Da Temperatura--------------")
print ("                                                 ")
celcius = float(input("digite a temperatura em graus celcius: "))
print ("                                                 ")

#Processamento
fahrenheit = (celcius * 9/5) + 32

#Saída
print("--------------------Conversão--------------------")
print ("                                                 ")
print(f"A temperatura em fahrenheit é: {fahrenheit}")
```
