# Escrever um algoritimo e o programa que receba 4 notas do aluno e calcula a média dessas notas e mostre a média

```python
# entrada

a = float (input ("digite a nota: "))
b = float (input ("digite a nota: "))
c = float (input ("digite a nota: "))
d = float (input ("digite a nota: "))

# processamento
média = (a + b + c + d) / 4
# saida
print (f"a média do aluno foi: {média}")
```
